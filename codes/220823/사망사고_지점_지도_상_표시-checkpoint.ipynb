```python
import folium
from folium.features import CustomIcon
import pandas as pd
```

```python
y_17 = pd.read_csv('./data/17년도분류.csv',encoding='cp949')
y_17.index = y_17['발생년월일시']
y_17.drop('Unnamed: 0', axis=1,inplace=True)
y_17
```

```python
def data_list(x):
    type_1 = x.loc[:,'당사자종별_1당'].to_list() # 특정 열 추출해서 리스트로 변환
    type_2 = x.loc[:,'당사자종별_2당'].to_list()
    type_acc = x.loc[:,'사고유형'].to_list()
    roads_detail = x.loc[:,'도로형태'].to_list()
    data_list = [type_1,type_2,type_acc,roads_detail]
    return data_list
```

```python
type_1 = data_list(y_17)[0]
type_2 = data_list(y_17)[1]
type_acc = data_list(y_17)[2]
roads_detail = data_list(y_17)[3]

accidents = []
for i in range(len(type_1)):
    if(type_1[i] == '개인형이동수단(PM)')|(type_2[i] == '개인형이동수단(PM)'):
        accidents.append(i)
        print(type_1[i],type_2[i])

coordinate = y_17.loc[:,['위도','경도']] # 특정 열 묶어서 데이터 프레임으로 추출

pm_dead = []
for x in accidents:
    pm_dead.append(coordinate.iloc[x].to_list())
    
string = []
for i in accidents:
    string.append('사고유형:' + str(type_acc[i]) + '<br>도로종류:' + str(roads_detail[i]))
string
```

```python
list_1 = []
list_2 = []
for i in accidents:
    list_1.append(type_acc[i])
    list_2.append(roads_detail[i])
df_1 = pd.DataFrame(list_1)
df_2 = pd.DataFrame(list_2)
details = pd.concat([df_1,df_2],axis=1)
details.columns = ['사고유형','도로종류']
df_tmp1 = pd.DataFrame(details['사고유형'].value_counts())
df_tmp2 = pd.DataFrame(details['도로종류'].value_counts())
df_tmp1
```

```python
df_tmp2
```
