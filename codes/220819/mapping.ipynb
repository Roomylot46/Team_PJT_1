```python
y_17 = pd.read_csv('./data/17년도분류.csv',encoding='cp949')
```

```python
y_17.index = y_17['발생년월일시']
```

```python
y_17.drop('Unnamed: 0', axis=1,inplace=True)
```

```python
y_17
```

```python
type_1 = y_17.loc[:,'당사자종별_1당'].to_list() # 특정 열 추출해서 리스트로 변환
type_2 = y_17.loc[:,'당사자종별_2당'].to_list()
type_acc = y_17.loc[:,'사고유형'].to_list()
roads_detail = y_17.loc[:,'도로형태'].to_list()
```

```python
accidents = []
for i in range(len(type_1)):
    if(type_1[i] == '개인형이동수단(PM)')|(type_2[i] == '개인형이동수단(PM)'):
        accidents.append(i)
        print(type_1[i],type_2[i])
accidents
```

```python
coordinate = y_17.loc[:,['위도','경도']] # 특정 열 묶어서 데이터 프레임으로 추출
```

```python
pm_dead = []
for x in accidents:
    pm_dead.append(coordinate.iloc[x].to_list())
pm_dead
```

```python
string = []
for i in accidents:
    string.append('사고유형:' + str(type_acc[i]) + '<br>도로종류:' + str(roads_detail[i]))
string
```

```python
mapping = folium.Map(location=pm_dead[0],zoom_start=16)
folium.Marker(pm_dead[0],
              tooltip=string[0],
              icon = folium.Icon(color='red', icon='exclamation-sign'),
              color='red',
              tiles='Stamenterrain').add_to(mapping)
mapping
```
