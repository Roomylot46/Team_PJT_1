```python
sum_17 = y_17.loc[:,['합계']]
sum_17.columns = ['단일로','교차로','기타']
sum_17
```

```python
del y_17['합계']
```

```python
y_17
```

```python
def customPct(pct) :
    return ('%.1f%%' % pct) if pct > 0 else ''

labels = sum_17.columns
labels_1 = y_17.columns
size = sum_17.iloc[0].to_list()
size_1 = y_17.iloc[0].to_list()
colors = ['darkred','chocolate','darkgoldenrod','darkslategrey','midnightblue','indigo','midiumvioletred']
colors_1 = ['salmon','lightsalmon','lemonchiffon','paleturquoise','cornflowerblue','mediumpurple','lightpink']

for i in range(len(sum_17.index)):
    size = sum_17.iloc[i].to_list()
    size_1 = y_17.iloc[i].to_list()
    colors=colors
    plt.figure(figsize=(10,10))
    plt.title(sum_17.index[i])

    pie_d = plt.pie(size_1, 
    #         labels=labels_1,
            colors = colors_1, 
            startangle=90, 
            autopct= customPct, 
            shadow=True, 
            textprops={'fontsize': 14},  
            pctdistance = 0.85,
            radius = 1)


    pie = plt.pie(size, 
            labels='<'+labels+'>', 
    #       explode=[0.05]*3, 
            colors = colors, 
            startangle=90, 
            autopct= customPct, 
            shadow=True, 
            textprops={'fontsize': 12},  
            pctdistance = 0.85,
            radius = 0.75,
            labeldistance=0.4)
    
    plt.legend(labels=labels_1,loc='best')
    centre_circle = plt.Circle((0,0),0.50,fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)
```
