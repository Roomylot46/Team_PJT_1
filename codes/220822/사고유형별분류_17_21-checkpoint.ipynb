```python
df_accident = df.loc['사고건수']
df_accident.index.name = '사고건수'
df_accident
```

```python
r_increase = df_accident.iloc[4].astype(int) / df_accident.iloc[0].astype(int)
df_increase = pd.DataFrame(r_increase)
df_increase
```

```python
accidents = df_accident['차대사람'].astype(int)
list_acc1 = []
for i in range(len(accidents)-1):
    rate = accidents[i+1] / accidents[i]
    list_acc1.append(rate)

accidents = df_accident['차대차'].astype(int)
list_acc2 = []
for i in range(len(accidents)-1):
    rate = accidents[i+1] / accidents[i]
    list_acc2.append(rate)

accidents = df_accident['차량단독'].astype(int)
list_acc3 = []
for i in range(len(accidents)-1):
    rate = accidents[i+1] / accidents[i]
    list_acc3.append(rate)
    
key = ['차대사람','차대차','차량단독']
list_acc = [list_acc1, list_acc2, list_acc3]
dict_data = dict(zip(key,list_acc))
df_acc = pd.DataFrame(dict_data)
df_acc.index = ['17-18 증가율', '18-19 증가율', '19-20 증가율', '20-21 증가율']
df_acc
```
