```python
def sel_column(column):
    list_1 = []
    for i in range(len(item)):
        columns = item[i].find_all('col')
        cols = []
        k = 0
        for j in range(len(columns)):
            if(columns[j]['name'] == column): # 필요한 column에 대한 index 값 가져오기
                k = j

        x = str(item[i]).split('/>')
        y = x[1:]
        tmp = []
        for j in range(len(y)):
            z = (y[j].split('<col')[0]).strip("</item>")
            tmp.append(z)
        list_1.append(tmp[k])
    return list_1
```
